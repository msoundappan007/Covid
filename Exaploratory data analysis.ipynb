# Total number of confirmed, death, recovered, active cases per WHO region

```python
agg_columns = {
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum',
    'Active': 'sum'
}
cases_per_region = df.groupby('WHO Region').agg(agg_columns).reset_index()
print(cases_per_region)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 7))
ax = sns.barplot(x='WHO Region', y='Confirmed', data=cases_per_region, palette='Set2')
ax.set_title('Total Confirmed Cases per WHO Region')
ax.set_xlabel('WHO Region')
ax.set_ylabel('Total Confirmed Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='WHO Region', y='Deaths', data=cases_per_region, palette='Set2')
ax.set_title('Total Death Cases per WHO Region')
ax.set_xlabel('WHO Region')
ax.set_ylabel('Total Death Cases')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='WHO Region', y='Recovered', data=cases_per_region, palette='Set2')
ax.set_title('Total Death Cases per WHO Region')
ax.set_xlabel('WHO Region')
ax.set_ylabel('Total Death Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='WHO Region', y='Active', data=cases_per_region, palette='Set2')
ax.set_title('Total Death Cases per WHO Region')
ax.set_xlabel('WHO Region')
ax.set_ylabel('Total Death Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Total number of confirmed,death, recovered, active cases per country

```python
agg_columns = {
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum',
    'Active': 'sum'
}
cases_per_country = df.groupby('Country/Region').agg(agg_columns).sort_values(by='Confirmed', ascending=False).reset_index()
print(cases_per_country)
```

```python
cases_per_country.head(10)
```

```python
cases_per_country=cases_per_country[0:10]
cases_per_country
```

```python
sns.set(style="whitegrid")

countries = cases_per_country['Country/Region']
confirmed_cases = cases_per_country['Confirmed']
death_cases = cases_per_country['Deaths']
recovered_cases = cases_per_country['Recovered']
active_cases = cases_per_country['Active']
data = pd.DataFrame({
    'Country': countries,
    'Confirmed': confirmed_cases,
    'Deaths': death_cases,
    'Recovered': recovered_cases,
    'Active': active_cases
})

# Melt the DataFrame for Seaborn
data_melted = data.melt(id_vars='Country', var_name='Case Type', value_name='Total Cases')

plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Total Cases', hue='Case Type', data=data_melted, palette='Set1')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.title('Total Cases per Country')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

# Which country of each WHO region has highest case.

```python
case_columns = ['Confirmed', 'Deaths', 'Recovered', 'Active']
highest_cases_per_region = pd.DataFrame(columns=['WHO Region'] + case_columns)
grouped = df.groupby('WHO Region')

for name, group in grouped:
    row_data = [name]
    for case_column in case_columns:
        country_with_highest_cases = group.loc[group[case_column].idxmax()]['Country/Region']
        row_data.append(country_with_highest_cases)
    highest_cases_per_region.loc[len(highest_cases_per_region)] = row_data

highest_cases_per_region

```

# recovery for top 10  country

```python
frame=df[['Country/Region','Recovered']].groupby('Country/Region').mean().sort_values(by='Recovered', ascending=False).reset_index()[0:10]
```

```python
countries = frame['Country/Region']
recovered_mean = frame['Recovered']
plt.figure(figsize=(10, 6))
plt.bar(countries, recovered_mean, color='green')
plt.xlabel('Country')
plt.ylabel('Average Recovered Cases')
plt.title('Top 10 Countries by Average Recovered Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
countries = frame['Country/Region']
recovered_mean = frame['Recovered']

# Set up colors using a color map
colors = plt.cm.Paired.colors
plt.figure(figsize=(8, 8))
plt.pie(recovered_mean, labels=countries, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Average Recovered Cases by Country')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

# % of recovery for each region

```python
df['Recovery Percentage'] = (df['Recovered'] / df['Confirmed']) * 100
recovery_per_region = df.groupby('WHO Region')['Recovery Percentage'].mean()
print(recovery_per_region)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

regions = recovery_per_region.index
recovery_percentages = recovery_per_region.values

# Set up colors using a color map
colors = plt.cm.Paired.colors
plt.figure(figsize=(8, 8))
plt.pie(recovery_percentages, labels=regions, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Average Recovery Percentages by Region')
plt.tight_layout()
plt.show()
```
